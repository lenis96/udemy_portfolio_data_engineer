```python
estaciones = df3[['Nombre_comercial','Direccion','Full_Address']].drop_duplicates(subset='Nombre_comercial')
estaciones.reset_index(inplace=True)
estaciones.drop(columns='index',inplace=True)
estaciones
```

```python
def get_coord(address,api_key):
    # https://developers.google.com/maps/documentation/geocoding/requests-geocoding?hl=en
    # https://developers.google.com/maps/documentation/geocoding/usage-and-billing?hl=es-419
    url = f'https://maps.googleapis.com/maps/api/geocode/json?address={address}&key={api_key}'
    try:

        res = requests.get(url)
        if res.status_code == 200:
            res = res.json()['results']
            if len(res) > 0:
                clean_address = res[0]['formatted_address']
                lat = res[0]['geometry']['location']['lat']
                lng = res[0]['geometry']['location']['lng']
                return [clean_address,lat,lng]
            else:
                return ['Not found','NA','NA']
        # print(res.status_code)
        # print(res.json())
    except Exception as e:
        return ['Not found','NA','NA']
        # print(e)

```

# Extracción de Cordenadas

```python
cordenadas = []
tiempo= [1,1.5,2]
# r = get_coord(estaciones['Full_Address'][0],API_KEY)
for i in tqdm(range(len(estaciones['Full_Address'])),colour = 'green'):
    try:
        r = get_coord(estaciones['Full_Address'][i],API_KEY)
        geo_source = r[1],r[2]
        cordenadas.append(geo_source)
    except:
        cordenadas.append('Error')
    time.sleep(random.choice(tiempo))

```

```python
estaciones['Coords'] = cordenadas
```

```python
estaciones
```

```python
df_final = pd.merge(df3,estaciones[['Full_Address','Coords']],how='inner',on='Full_Address')
```
